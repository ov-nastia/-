```python
import numpy as np
```

```python
#ЗАВДАННЯ 1

#Створіть одновимірний масив (вектор) з простими числами, не перевищуючи 70, та виведіть його значення.

# Створюємо одновимірний масив простих чисел до 70
vector = np.array([num for num in range(2, 71) if all(num % i != 0 for i in range(2, int(np.sqrt(num)) + 1))])

# Виводимо масив простих чисел
print("Масив простих чисел до 70:")
print(vector)

```

```python
#ЗАВДАННЯ 11

#Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму елементів матриці, які менше за відповідний елемент вектору.

matrix = np.random.rand(2, 3)
vector = np.random.rand(3, 1)

comparison_matrix = matrix < vector.T
sum_elements = np.sum(matrix[comparison_matrix])

print("Матриця 2x3:")
print(matrix)
print("\nВектор 3x1:")
print(vector)
print("\nСума елементів матриці, які менші за відповідний елемент вектору:")
print(sum_elements)
```

```python
#ЗАВДАННЯ 14

#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -5 до 5 та знайдіть добуток елементів матриці, які більше за їх середнє арифметичне значення

matrix = np.random.uniform(-5, 5, size=(3, 3))
print("Матриця:\n", matrix)

mean_value = np.mean(matrix)
print("Середнє арифметичне:", mean_value)

elements = matrix[matrix > mean_value]
product = np.prod(elements) if elements.size > 0 else 0
print("Добуток елементів більших за середнє:", product)
```

```python
# Створення матриці 3x3 з випадковими числами
matrix = np.random.uniform(-0.5, 0.5, size=(3, 3))
print("Матриця:\n", matrix)


def find_max_int(row):
    rounded_row = np.round(row)
    max_index = np.argmax(np.abs(rounded_row))
    return rounded_row[max_index]

result_vector = np.apply_along_axis(find_max_int, 1, matrix).reshape(-1, 1)

print("Вектор-стовпчик з найбільшими за модулем цілими значеннями:\n", result_vector)
```
