3. Побудова і налаштування моделей:

```python
#3.1 Розділіть дані на навчальний і тестовий набори.
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Масштабування даних
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#3.2. Побудуйте три моделі класифікації (на власний вибір),наприклад:
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}

param_grids = {
    'Logistic Regression': {
        'C': [0.1, 1, 10, 100],
        'penalty': ['l2']
    },
    'Decision Tree': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_leaf': [1, 3, 5]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30]
    }
}
```

```python
#3.3 Підберіть оптимальні параметри для кожної моделі за допомогою GridSearchCV (можна вибрати інший оптимізатор).
def tune_and_train_model(model_name, model, param_grid):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_params = grid_search.best_params_
    best_model = grid_search.best_estimator_
    return best_model, best_params

# Навчання і підбір гіперпараметрів для кожної моделі
best_models = {}
best_params = {}

for model_name, model in models.items():
    print(f"\nТонування моделі {model_name}...")
    best_model, params = tune_and_train_model(model_name, model, param_grids[model_name])
    best_models[model_name] = best_model
    best_params[model_name] = params

# Виведення найкращих параметрів
print("\nНайкращі параметри для кожної моделі:")
for model_name, params in best_params.items():
    print(f"{model_name}: {params}")
```
